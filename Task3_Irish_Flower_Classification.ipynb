Applying Elbow Tecnique

```python
k_rng = range(1, 10)
sse = []
for k in k_rng:
  km= KMeans(n_clusters=k) 
  km.fit(iris_flower[['petal_length', 'petal_width']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k_rng')
plt.ylabel('sum of squared errors')
plt.plot(k_rng , sse)
```

# Applying KMean Algorithm

```python
km =KMeans(n_clusters=3 , random_state=0)
y_predicted =km.fit_predict(iris_flower[['petal_length' , 'petal_width']])
y_predicted
```

```python
iris_flower['cluster']=y_predicted
iris_flower.head(150)
```

# Accuracy Measure

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(iris_flower.species , iris_flower.cluster)
cm
```

```python
true_labels = iris_flower.species
predict_labels = iris_flower.cluster


cm = confusion_matrix(true_labels ,predict_labels)
class_labels = ['Setosa' ,'versicolor' ,'verginica']


#plot Confusion Metrix
plt.imshow(cm , interpolation='nearest' , cmap=plt.cm.Blues)
plt.title('Confusion Metrix')
plt.colorbar()
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks ,class_labels)
plt.yticks(tick_marks ,class_labels)


# Fill Metrix with values
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j,i , str(cm[i][j]), ha='center', va ='center' , color='white')
        
plt.xlabel('Predict Label')
plt.ylabel('True Label')
plt.show
```
